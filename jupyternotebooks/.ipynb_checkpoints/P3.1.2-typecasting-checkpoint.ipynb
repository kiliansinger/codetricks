### Fortran

```fortran
program main
implicit none
integer :: stat
integer :: i
real*8 :: d
character*20 :: s
character*30 :: s2
print *,7/8
print *,int(7.8d0)
i=7;d=7.2
write(s,"(i5)") i
print *,s
write(s,"(f10.5)") d
print *,s
write(s,"(es20.5)") d
print *,s
i=2;d=2.1
read(s,*,iostat=stat) i
read(s2,*,iostat=stat) d
print *,i,d
!print *,logical(1)
!print *,logical(0)
end program
```
