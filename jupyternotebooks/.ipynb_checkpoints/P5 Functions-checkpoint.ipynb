### c++

````python
%%cpp -d
double square(double d){
    return d*d;
}
int fib(int x) {
   if((x==1)||(x==0)) {
      return(x);
   }else {
      return(fib(x-1)+fib(x-2));
   }
}
````

````python
%%cpp
cout <<square(2)<<" "<<fib(30)<<endl;
````

### Fortran

```fortran
program helloworld
implicit none
integer :: fib
real*8::square

print *,square(7.8d0)
print *,fib(10)
call test(5d0)
read *
end program

  
real*8 function square(d)
    implicit none
    real*8,intent(in)::d  ! intent can be "in", "out" or "inout"
    square=d*d
end function

subroutine test(d)
    implicit none
    real*8,intent(in)::d
    print *,d
end subroutine

pure recursive integer function fib(x) ! pure make no global variables accessible
    implicit none
    integer,intent(in)::x
    if (x.eq.1 .or. x.eq.0) then
        fib=x
        return
    else
        fib=fib(x-1)+fib(x-2)
        return
    endif
end function
```
