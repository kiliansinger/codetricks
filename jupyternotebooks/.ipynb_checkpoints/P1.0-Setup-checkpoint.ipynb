## Setup of C/C++ in Jupyterlab

Download Jupyterlab for your platform from: https://github.com/jupyterlab/jupyterlab-desktop
and instal.
Confirmed working on: Version 4.2.5 (also has my patch to work better on windows)

download
https://root.cern/install/all_releases/
and install
confirmed working on:Release 6.34.00.

open in jupyterlab the folder pane and select the + button and open a terminal.
type: where.exe conda.exe
%APPDATA%\jupyterlab-desktop\jlab_server\Scripts\conda.exe
and use it in a admin command promt that you open by typing win+s and typing cmd and then select 
run as administrator.

 you need to install python3.11 in this admin command prompt:
```bash
%APPDATA%\jupyterlab-desktop\jlab_server\Scripts\conda.exe init
%APPDATA%\jupyterlab-desktop\jlab_server\Scripts\conda.exe create --name py311 python=3.11
#close the shell and reopen it:
%APPDATA%\jupyterlab-desktop\jlab_server\Scripts\conda.exe activate py311
pip install ipykernel
python -m ipykernel install --user --name py311
python -m ipykernel install --user --name py311 --display-name "Python (py311)"
```
